```python
import pandas as pd
```

```python
df = pd.read_csv("/tmp/stats.csv", )
import matplotlib
import matplotlib.pyplot as plt
# matplotlib.style.use('ggplot')
```

```python
% matplotlib inline
plt.figure()
plt.ylabel("Probability")
plt.ylim([0,1])
df["driver_p"].plot(legend=True)
df["worker_p"].plot(legend=True)
plt.xlabel("Iterations")
# # df.plot("worker_p")
```

```python
plt.figure()
plt.ylabel("Time (ms)")
plt.ylim([0,0.1])
df.where(df["type"] == "driver")["time"][10:].plot(style='o', label="Driver")
df.where(df["type"] == "worker")["time"][10:].plot(style='x', label="Worker")
plt.xlabel("Iterations")
plt.legend()
```
