```python
import pandas as pd
```

```python
df = pd.read_csv("../stats.csv")[90:]
import matplotlib
import matplotlib.pyplot as plt
# matplotlib.style.use('ggplot')
```

```python
% matplotlib inline
plt.figure()
plt.ylabel("Probability")
plt.ylim([0,1])
df["driver_p"].plot(legend=True, label="Driver")
df["worker_p"].plot(legend=True, label="Worker")
plt.xlabel("Iterations")
plt.savefig("./sgd_gpu_prob.pdf")
# # df.plot("worker_p")
```

```python
plt.figure()
plt.ylabel("Time (seconds)")
plt.ylim([0,0.1])
df.where(df["type"] == "driver")["time"].plot(style='o', label="Driver")
df.where(df["type"] == "worker")["time"].plot(style='x', label="Worker")
plt.xlabel("Iterations")
plt.legend()
plt.savefig("./sgd_gpu_timing.pdf")
```
