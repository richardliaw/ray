```python
import pandas as pd
```

```python
df = pd.read_csv("stats.csv")
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
% matplotlib inline
plt.figure()
plt.ylabel("Probability")
df["driver_p"].plot(legend=True)
df["worker_p"].plot(legend=True)
plt.xlabel("Iterations")
# df.plot("worker_p")
```

```python
plt.figure()
plt.ylabel("Time (ms)")
df.where(df["type"] == "driver")["time"][10:].plot(style='x', legend=True, label="Driver")
df.where(df["type"] == "worker")["time"][10:].plot(style='o', legend=True, label="Worker")
plt.xlabel("Iterations")
```
