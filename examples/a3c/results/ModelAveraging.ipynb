```
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
from datetime import timedelta
plt.style.use('ggplot')
% matplotlib inline
import os
```

```
def load_dataframe(fname):
    with open(fname) as f:
        raw = json.load(f)
    raw["mean"] = [x[0][0] for x in raw['stats']]
    raw["std"] = [x[0][1] for x in raw['stats']]
    df = pd.DataFrame(data=raw)
    return df
```

```
paths = [os.path.join(x[0], y) for x in os.walk("./1_3/") for y in x[-1] ]
for fname in paths:
    df = load_dataframe(fname)
    plt.plot(df['TS'], df['mean'].rolling(5, ).mean())
    print(str(timedelta(seconds=df["TS"].max())))
plt.xlim([0, 180])
```
